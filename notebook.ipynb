```python
import sqlite3
```

```python
conn = sqlite3.connect("ringana.db")
```

```python
# create table with description, first_name, last_name, customer_number, email, country, salutation

stmt = """
CREATE TABLE IF NOT EXISTS users (
    description TEXT,
    first_name TEXT,
    last_name TEXT,
    customer_number TEXT,
    email TEXT,
    country TEXT,
    salutation TEXT
)
"""

conn.execute(stmt)
```

```python
# get schema

stmt = "SELECT sql FROM sqlite_master WHERE name='users'"
```

```python
cur = conn.execute(stmt)
```

```python
for r in cur.fetchall():
    print(r)
```
